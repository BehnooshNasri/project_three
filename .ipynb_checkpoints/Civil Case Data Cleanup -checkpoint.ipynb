```python
# Import dependencies
import pandas as pd 
```

```python
# Read the CSV data into a Pandas DataFrame
starter_df = pd.read_csv('../data/civil_cases.csv')
starter_df.head(10)
```

```python
starter_df.columns
```

```python
starter_df.dtypes
```

```python
civil_df = starter_df[['REF_DATE', 'GEO', 'Level of court', 'Type of case','VALUE']]
civil_df = civil_df.rename(columns={'REF_DATE': "Year_Period", 'GEO': "Jurisdiction",'Type of case': "Case_Type", 'VALUE': "Case_Count"})
civil_df.head(10)
```

```python
civil_df['Level of court'].unique()
```

```python
civil_df = civil_df.groupby(["Year_Period", "Jurisdiction", "Case_Type"])["Case_Count"].sum().reset_index()
civil_df
```

```python
civil_df["Case_Count"] = civil_df["Case_Count"].astype(int)
civil_df
```

```python
# Save the DataFrame to a CSV file
civil_df.to_csv('civil_case_data.csv', index=False)
```
